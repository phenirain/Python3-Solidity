```python
# Цель: на данной практической работе необходимо использовать датасет с прошлой практической работы и визуализировать данные в виде графиков

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fn = pd.read_csv("Financial Statements.csv")
plt.figure(figsize=(10, 6))
plt.plot(fn.loc[fn["Company "] == 'AAPL']["Year"], fn.loc[fn["Company "] == 'AAPL']["Gross Profit"], label="APPL")
plt.plot(fn.loc[fn["Company "] == 'MSFT']["Year"], fn.loc[fn["Company "] == 'MSFT']["Gross Profit"],
         label="MSFT")
plt.plot(fn.loc[fn["Company "] == 'GOOG']["Year"], fn.loc[fn["Company "] == 'GOOG']["Gross Profit"],
         label="GOOG")
plt.title("APPL profit")
plt.xlabel("year")
plt.ylabel("profit ($)")
plt.legend()
plt.show()
# с помощью loc мы фильтруем колонны по названию компаний берем значения колонн года и профита 
# по y мы показываем профит в $, по x мы показываем года
# каждая линия графика отражает отдельную компанию
```

```python
plt.figure(figsize=(10, 6))
plt.bar(fn.loc[fn["Company "] == "PYPL"]["Year"], fn.loc[fn["Company "] == "PYPL"]["ROI"], label="PYPL")
plt.bar(fn.loc[fn["Company "] == "AIG"]["Year"], fn.loc[fn["Company "] == "AIG"]["ROI"], label="AIG")
plt.bar(fn.loc[fn["Company "] == "PCG"]["Year"], fn.loc[fn["Company "] == "PCG"]["ROI"], label="PCG")
plt.title("ROI Companies")
plt.xlabel("Year")
plt.ylabel("ROI %")
plt.legend()
plt.show()
# мы создаем столбчатые графики по названиям компаний берем значения колонн года и окупаемость вложений
# компании PYPL AIG PCG
# создаем название, объявляем подпись для x и y, делаем легенду и показываем таблицу
# каждая линия графика отражает отдельную компанию
```

```python
plt.figure(figsize=(8, 8))
plt.pie([fn.loc[fn["Company "] == "PYPL"]["Current Ratio"].mean(),
         fn.loc[fn["Company "] == "MSFT"]["Current Ratio"].mean(),
         fn.loc[fn["Company "] == "AIG"]["Current Ratio"].mean(),
         fn.loc[fn["Company "] == "PCG"]["Current Ratio"].mean()], startangle=90, labels=["PYPL", "MSFT", "AIG", "PCG"])
plt.title("Current Ratio")
plt.show()
# тут мы создаем круговую диаграмму по разным компания получая их при помощи loc и получая их среднее значения
# за все года по коэффиценту ликвидности
# каждому куску соответсвует каждая компания
```

```python
plt.figure(figsize=(8, 8))
plt.pie([len(fn.loc[fn["Category"] == "IT"]),
         len(fn.loc[fn["Category"] == "Bank"]),
         len(fn.loc[fn["Category"] == "FinTech"]),
         len(fn.loc[fn["Category"] == "Bank"])], startangle=90, labels=["IT", "Bank", "FinTech", "Bank"])
plt.title("Categories")
plt.show()
# тут мы создаем круговую диаграмму по разным компания получая их при помощи loc и названия категорий
# и подсчитывая их создаем диаграмму
# каждому куску соответсвует каждая компания
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(fn.loc[fn["Category"] == "IT"]["Current Ratio"], fn.loc[fn["Category"] == "IT"]["Gross Profit"])
plt.title('Current Ratio to Gross Profit IT category')
plt.xlabel('current ratio')
plt.ylabel('gross profit')
plt.show()
# получаем записи по колонне IT через loc получаем коэффицент ликвидности и профит и по этим данным
# по их соотношению строим диаграмму, указываем название диаграммы и подписываем x и y
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(fn.loc[fn["Category"] == "IT"]["Gross Profit"], fn.loc[fn["Category"] == "IT"]["ROI"])
plt.title('Gross Profit to ROI IT category')
plt.xlabel('current ratio')
plt.ylabel('gross profit')
plt.show()
# получаем записи по колонне IT через loc получаем профит и коэффицент окупаемости и по этим данным
# по их соотношению строим диаграмму, указываем название диаграммы и подписываем x и y
```

```python

plt.figure(figsize=(12, 6))
plt.hist(fn["Category"], bins=20, alpha=0.7, label='категории')
plt.title('категории')
plt.xlabel('названаия')
plt.ylabel('количество')
plt.legend()
plt.show()
# мы формируем гистограмму по количеству категорий в дата-сете
# подписываем название, подписываем x и y
```

```python
plt.figure(figsize=(12, 8))
plt.hist(fn.loc[fn["Category"] == "IT"]["Current Ratio"], bins=20, alpha=0.7, label='Коэффицент ликвидности')
plt.title('Коэффицент ликвидности')
plt.ylabel('процент')
plt.legend()
plt.show()
# мы формируем гистограмму по коэффиценту ликвидности в дата-сете
# подписываем название и y
```

```python
plt.figure(figsize=(12, 8))
plt.hist(fn.loc[fn["Category"] == "IT"]["Gross Profit"], bins=20, alpha=0.7, label='Профит')
plt.title('Профит')
plt.ylabel('доход в $')
plt.legend()
plt.show()
# мы формируем гистограмму по профиту компаний в дата-сете
# подписываем название и y
```
