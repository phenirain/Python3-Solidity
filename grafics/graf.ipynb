```python
# :                 

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fn = pd.read_csv("Financial Statements.csv")
plt.figure(figsize=(10, 6))
plt.plot(fn.loc[fn["Company "] == 'AAPL']["Year"], fn.loc[fn["Company "] == 'AAPL']["Gross Profit"], label="APPL")
plt.plot(fn.loc[fn["Company "] == 'MSFT']["Year"], fn.loc[fn["Company "] == 'MSFT']["Gross Profit"],
         label="MSFT")
plt.plot(fn.loc[fn["Company "] == 'GOOG']["Year"], fn.loc[fn["Company "] == 'GOOG']["Gross Profit"],
         label="GOOG")
plt.title("APPL profit")
plt.xlabel("year")
plt.ylabel("profit ($)")
plt.legend()
plt.show()
#   loc             
#  y     $,  x   
#      
```

```python
plt.figure(figsize=(10, 6))
plt.bar(fn.loc[fn["Company "] == "PYPL"]["Year"], fn.loc[fn["Company "] == "PYPL"]["ROI"], label="PYPL")
plt.bar(fn.loc[fn["Company "] == "AIG"]["Year"], fn.loc[fn["Company "] == "AIG"]["ROI"], label="AIG")
plt.bar(fn.loc[fn["Company "] == "PCG"]["Year"], fn.loc[fn["Company "] == "PCG"]["ROI"], label="PCG")
plt.title("ROI Companies")
plt.xlabel("Year")
plt.ylabel("ROI %")
plt.legend()
plt.show()
#              
#  PYPL AIG PCG
#  ,    x  y,     
#      
```

```python
plt.figure(figsize=(8, 8))
plt.pie([fn.loc[fn["Company "] == "PYPL"]["Current Ratio"].mean(),
         fn.loc[fn["Company "] == "MSFT"]["Current Ratio"].mean(),
         fn.loc[fn["Company "] == "AIG"]["Current Ratio"].mean(),
         fn.loc[fn["Company "] == "PCG"]["Current Ratio"].mean()], startangle=90, labels=["PYPL", "MSFT", "AIG", "PCG"])
plt.title("Current Ratio")
plt.show()
#             loc     
#      
#     
```

```python
plt.figure(figsize=(8, 8))
plt.pie([len(fn.loc[fn["Category"] == "IT"]),
         len(fn.loc[fn["Category"] == "Bank"]),
         len(fn.loc[fn["Category"] == "FinTech"]),
         len(fn.loc[fn["Category"] == "Bank"])], startangle=90, labels=["IT", "Bank", "FinTech", "Bank"])
plt.title("Categories")
plt.show()
#             loc   
#     
#     
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(fn.loc[fn["Category"] == "IT"]["Current Ratio"], fn.loc[fn["Category"] == "IT"]["Gross Profit"])
plt.title('Current Ratio to Gross Profit IT category')
plt.xlabel('current ratio')
plt.ylabel('gross profit')
plt.show()
#     IT  loc         
#     ,      x  y
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(fn.loc[fn["Category"] == "IT"]["Gross Profit"], fn.loc[fn["Category"] == "IT"]["ROI"])
plt.title('Gross Profit to ROI IT category')
plt.xlabel('current ratio')
plt.ylabel('gross profit')
plt.show()
#     IT  loc         
#     ,      x  y
```

```python

plt.figure(figsize=(12, 6))
plt.hist(fn["Category"], bins=20, alpha=0.7, label='')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.legend()
plt.show()
#        -
#  ,  x  y
```
